```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np


wcss = []  
silhouette_scores = []  
k_range = range(2, 11)  

print("Запускаем метод локтя для K от 2 до 10...")

for k in k_range:
    
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    
    
    wcss.append(kmeans.inertia_)
    
    
    silhouette_avg = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)
    
    print(f"K={k}: WCSS = {kmeans.inertia_:.2f}, Silhouette = {silhouette_avg:.3f}")


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))


ax1.plot(k_range, wcss, 'bo-', markersize=8, linewidth=2)
ax1.set_xlabel('Количество кластеров (K)')
ax1.set_ylabel('Within-Cluster Sum of Square (WCSS)')
ax1.set_title('Метод локтя для оптимального K')
ax1.grid(True, alpha=0.3)


for i, (k, w) in enumerate(zip(k_range, wcss)):
    ax1.annotate(f'K={k}', (k, w), textcoords="offset points", xytext=(0,10), ha='center')


ax2.plot(k_range, silhouette_scores, 'ro-', markersize=8, linewidth=2)
ax2.set_xlabel('Количество кластеров (K)')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Метрика силуэта для разных K')
ax2.grid(True, alpha=0.3)


best_k_silhouette = k_range[np.argmax(silhouette_scores)]
best_silhouette = max(silhouette_scores)
ax2.axvline(x=best_k_silhouette, color='red', linestyle='--', alpha=0.7, 
            label=f'Лучшее K={best_k_silhouette}\nSilhouette={best_silhouette:.3f}')
ax2.legend()

plt.tight_layout()
plt.show()

# Анализ результатов
print("\n АНАЛИЗ РЕЗУЛЬТАТОВ:")
print("=" * 50)


wcss_diff = [wcss[i-1] - wcss[i] for i in range(1, len(wcss))]
elbow_k = k_range[np.argmax(wcss_diff) + 1]  

print(f" Рекомендуемое число кластеров по методу локтя: K = {elbow_k}")
print(f" Лучшее число кластеров по метрике силуэта: K = {best_k_silhouette}")
print(f" Лучший silhouette score: {best_silhouette:.3f}")


if best_silhouette > 0.7:
    print(" Отличное разделение кластеров!")
elif best_silhouette > 0.5:
    print("  Умеренное разделение кластеров")
elif best_silhouette > 0.25:
    print(" Слабое разделение кластеров")
else:
    print(" Очень слабое разделение, кластеры нечеткие")

print(f"\n Совет: Попробуй K = {elbow_k} или K = {best_k_silhouette} для финальной кластеризации")
```

```python

df_clean_20k_no_outliers['kmeans_cluster'] = kmeans_labels

def create_hotel_cluster_profiles(df, cluster_column):
    
    profiles = {}
    
    for cluster in sorted(df[cluster_column].unique()):
        cluster_data = df[df[cluster_column] == cluster]
        
        profile = {
            
            'size': len(cluster_data),
            'size_percentage': len(cluster_data) / len(df) * 100,
            
            
            'avg_lead_time': cluster_data['lead_time'].mean(),
            'avg_adr': cluster_data['adr'].mean(),
            'avg_stay_length': (cluster_data['stays_in_weekend_nights'] + cluster_data['stays_in_week_nights']).mean(),
            'avg_guests': cluster_data['adults'].mean() + cluster_data['children'].mean(),
            
            
            'avg_special_requests': cluster_data['total_of_special_requests'].mean(),
            'avg_booking_changes': cluster_data['booking_changes'].mean(),
            'previous_cancellations_rate': cluster_data['previous_cancellations'].mean(),
            
            
            'parking_need': cluster_data['required_car_parking_spaces'].mean(),
            
            
            'client_profile': determine_client_profile(cluster_data)
        }
        
        
        categorical_cols = ['hotel', 'meal', 'market_segment', 'customer_type', 'deposit_type']
        for cat_col in categorical_cols:
            if cat_col in df.columns:
                most_common = cluster_data[cat_col].mode()
                profile[f'most_common_{cat_col}'] = most_common[0] if len(most_common) > 0 else 'N/A'
                # Процент самого частого значения
                value_counts = cluster_data[cat_col].value_counts(normalize=True)
                profile[f'percent_{cat_col}'] = value_counts.iloc[0] * 100 if len(value_counts) > 0 else 0
        
        profiles[f'Cluster_{cluster}'] = profile
    
    
    profiles_df = pd.DataFrame(profiles).T
    
    
    print(" ПРОФИЛИ КЛАСТЕРОВ КЛИЕНТОВ ОТЕЛЯ")
    print("=" * 60)
    
    for cluster, profile in profiles.items():
        print(f"\n{cluster}:")
        print(f"   Размер: {profile['size']} клиентов ({profile['size_percentage']:.1f}%)")
        print(f"   Профиль: {profile['client_profile']}")
        print(f"   Средняя цена: ${profile['avg_adr']:.2f}")
        print(f"   Время бронирования: {profile['avg_lead_time']:.1f} дней")
        print(f"   Длина проживания: {profile['avg_stay_length']:.1f} ночей")
        print(f"   Гостей: {profile['avg_guests']:.1f}")
        print(f"   Особых запросов: {profile['avg_special_requests']:.1f}")
        print(f"   Изменений брони: {profile['avg_booking_changes']:.1f}")
        
        
        print(f"    Отель: {profile.get('most_common_hotel', 'N/A')}")
        print(f"    Питание: {profile.get('most_common_meal', 'N/A')}")
        print(f"   Сегмент: {profile.get('most_common_market_segment', 'N/A')}")
        print(f"   Тип клиента: {profile.get('most_common_customer_type', 'N/A')}")
        
        if profile['previous_cancellations_rate'] > 0.5:
            print(f"    Высокий риск отмены!")
    
    return profiles_df


def determine_client_profile(cluster_data):
    avg_lead_time = cluster_data['lead_time'].mean()
    avg_adr = cluster_data['adr'].mean()
    avg_guests = cluster_data['adults'].mean() + cluster_data['children'].mean()
    
    if avg_adr > 100 and avg_lead_time < 30:
        return "Бизнес-путешественники"
    elif avg_lead_time > 100:
        return "Планировщики отпуска"
    elif avg_guests > 2.5:
        return "Семейные отдыхающие"
    elif cluster_data['total_of_special_requests'].mean() > 2:
        return "Требовательные клиенты"
    else:
        return "Стандартные туристы"


print("Создаем профили кластеров...")
kmeans_profiles = create_hotel_cluster_profiles(df_clean_20k_no_outliers, 'kmeans_cluster')


print("\n СРАВНЕНИЕ КЛАСТЕРОВ ПО КЛЮЧЕВЫМ ПАРАМЕТРАМ:")
key_metrics = ['lead_time', 'adr', 'total_of_special_requests', 'stays_in_week_nights']
cluster_means = df_clean_20k_no_outliers.groupby('kmeans_cluster')[key_metrics].mean()
print(cluster_means)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import numpy as np


df_with_clusters = df_clean_20k_no_outliers.copy()


eps_values = np.arange(0.5, 3.1, 0.5)
min_samples_values = [3, 5, 7, 10]

best_sil = -1
best_params = None

for eps in eps_values:
    for min_samples in min_samples_values:
        db = DBSCAN(eps=eps, min_samples=min_samples)
        labels = db.fit_predict(X)  # Используем X (уже масштабированные данные)
        
        # Нужно минимум 2 кластера, чтобы посчитать silhouette
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        if n_clusters >= 2:
            sil = silhouette_score(X, labels)
            n_noise = list(labels).count(-1)
            print(f"eps={eps:.1f}, min_samples={min_samples}, кластеры={n_clusters}, шум={n_noise}, silhouette={sil:.4f}")
            if sil > best_sil:
                best_sil = sil
                best_params = (eps, min_samples, n_clusters)
        else:
            n_noise = list(labels).count(-1)
            print(f"eps={eps:.1f}, min_samples={min_samples}, кластеры={n_clusters}, шум={n_noise} - недостаточно кластеров")

# Запускаем с лучшими параметрами
if best_params:
    print(f"\n ЛУЧШИЕ ПАРАМЕТРЫ: eps={best_params[0]}, min_samples={best_params[1]}")
    db_best = DBSCAN(eps=best_params[0], min_samples=best_params[1])
    dbscan_labels = db_best.fit_predict(X)
    
    df_with_clusters['dbscan_cluster'] = dbscan_labels
    
    # Финальные метрики
    n_clusters_dbscan = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
    n_noise = list(dbscan_labels).count(-1)
    
    print(f"\n ФИНАЛЬНЫЕ РЕЗУЛЬТАТЫ DBSCAN:")
    print(f"Количество кластеров: {n_clusters_dbscan}")
    print(f"Количество точек шума: {n_noise}")
    
    if n_clusters_dbscan > 1:
        dbscan_silhouette = silhouette_score(X, dbscan_labels)
        dbscan_calinski = calinski_harabasz_score(X, dbscan_labels)
        dbscan_davies = davies_bouldin_score(X, dbscan_labels)

        print(f"Silhouette Score: {dbscan_silhouette:.3f}")
        print(f"Calinski-Harabasz Score: {dbscan_calinski:.3f}")
        print(f"Davies-Bouldin Score: {dbscan_davies:.3f}")
else:
    print("Не удалось найти параметры, дающие хотя бы 2 кластера")
```
